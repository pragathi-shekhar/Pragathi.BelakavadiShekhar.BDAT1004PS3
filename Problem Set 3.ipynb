```python
# Q4 step 1, 2 and 3

import pandas as pd
import datetime
data = pd.read_csv(r'wind.txt',sep='\s+',index_col=False,parse_dates = {'Date':['Yr','Mo','Dy']})
data.head()
```

```python
#Q4 step 4

def correct_year(val):
    year = val.year - 100  if val.year>2018 else val.year
    return datetime.date(year, val.month, val.day)
    
data['Date'] = data['Date'].apply(correct_year)
data
```

```python
#Q4 Step5

data = data.set_index('Date')
data.index.astype('datetime64[ns]')
```

```python
#Q4 step 6 - missing values count

print(data.iloc[1:].isnull().sum())
```

```python
#Q4 step 7 - non-missing values count

print(data.count().sum())
```

```python
#Q4 step 8 - mean windspeeds of the windspeeds over all the locations and

mean_windspeed = data.mean()
mean_windspeed.mean()
```

```python
#Q4 step 9 

loc_stats = data.agg(['min','max','mean','std'])
loc_stats
```

```python
#Q4 step 10

day_stats = pd.DataFrame()

day_stats['min'] = data.min(axis=1)
day_stats['max'] = data.max(axis = 1) 
day_stats['mean'] = data.mean(axis = 1) 
day_stats['std'] = data.std(axis = 1) 
day_stats
```

```python
#Q4 step 11

data.index.astype('datetime64[ns]')
#print([pd.DatetimeIndex(data['Date']).month])

#print(data['Date'].dt.month)
```

```python
#Q4 step 12 

yearly = data.resample('A').mean()
yearly
```

```python
#Q4 step 13

monthly = data.resample('M').mean()
monthly
```

```python
#Q4 step 14

weekly = data.resample('W').mean()
weekly
```

```python
#Q4 step 15

weekly = data.resample('W').agg(['min','max','mean','std'])
weekly
```

```python
#Q7 - Marriages and Divorces per capita in US

import pandas as pd #import pandas
import matplotlib.pyplot as plt 

marriage_divorce = pd.read_csv('us-marriages-divorces-1867-2014.csv') #read csv into dataframe 

marriage_divorce = marriage_divorce[marriage_divorce['Year'].apply(lambda x: x in [1900, 1950, 2000])] #apply filter on year
data = pd.DataFrame(marriage_divorce, columns=["Marriages_per_1000","Divorces_per_1000","Year"])
 
# plot the dataframe
data.plot(x="Year", y=["Marriages_per_1000","Divorces_per_1000"], kind="bar", figsize=(9, 8))
 
# print bar graph
plt.show()
```
